# Project Description
Mushrooms are delicious but can also be dangerous. Using data on mushrooms I will try to predict whether it is poisonous or not.

# Data Preprocessing
## Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the Dataset

```python
dataset = pd.read_csv('mushrooms.csv')
```

## Showing the Dataset

```python
pd.DataFrame(dataset)
```

## Seperate The Input and Output
Here, we put the independent variables in X and the dependent variable in y. 

```python
X = dataset.iloc[:, 1:]
Y = dataset.iloc[:, 0:1]
```

## Encoding Categorical Data
One hot encoding the independent variables.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# instantiate OneHotEncoder object
ohe = OneHotEncoder()
# instantiate ColumnTransformer object
ct = ColumnTransformer(transformers=[('encoder1', ohe, list(X.columns))], remainder='passthrough')
# transform data in array X
X = ohe.fit_transform(X)
X
```

Label encoding dependant variable.

```python
from sklearn.preprocessing import LabelEncoder

# instantiate LabelEncoder object
le = LabelEncoder()
# encode data
Y = pd.DataFrame(Y)
Y['class'] = le.fit_transform(Y['class'])
Y = np.asarray(Y)
```

## Splitting the Dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split

# split dataset, train:test split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

# Classification Model Analysis
## KNN
Train the classification model.

```python
from sklearn.neighbors import KNeighborsClassifier 

# instantiate knn object
knn = KNeighborsClassifier(n_neighbors=3)
# train model
knn = knn.fit(X_train, np.ravel(y_train))
knn.score(X_test, np.ravel(y_test))
```

# Conclusion
The models accuracy using KNN with 3 neighbors is 100%!
